```python
import pandas as pd
import numpy as np
import pylab as pl
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
%matplotlib inline
sns.set_style('whitegrid')
```

```python
from numpy.random import randn
data = randn(50)
```

```python
data
```

```python
plt.hist(data, alpha = 0.5)
sns.rugplot(data)
```

```python
sns.kdeplot(data)
sns.rugplot(data)
```

```python
sns.rugplot(data)
list = [0.1,0.3,0.5]

for num in list:
    sns.kdeplot(data, bw = num, label = num)
```

```python
sns.rugplot(data)
options = ['biw', 'tri','gau']

for option in options:
    sns.kdeplot(data, kernel = option, label = option, bw = 0.5)
```

```python
iris = sns.load_dataset("iris")
sns.kdeplot(iris.sepal_length, iris.petal_width, shade = True)
```

```python
sns.jointplot(x = 'sepal_length', y = 'petal_width', kind = 'kde', data = iris)
```
