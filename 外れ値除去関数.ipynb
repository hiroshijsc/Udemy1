# このデータセットだとx６が一番精度が高い

```python
def outlier_removal(df):
    col = 'x6'
    mean = df.mean()
    sigma = df.std()
    low = mean[col] - 3*sigma[col]
    high = mean[col] + 3*sigma[col]
    outlier_removal_df =df[(low<df[col]) & (df[col]<high)]
    return outlier_removal_df
```
